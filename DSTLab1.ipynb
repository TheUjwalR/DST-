```python
#Class to create a Node
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None

#Class to create a Linked List and perform operations on it
class LinkedList:
  def __init__(self):
    self.head=None

  #Print the Linked List
  def printLinkedList(self):
    if self.head==None:
      print("The Linked List is empty!")
    else:
      current_node=self.head
      while current_node!=None:
        print(f" {current_node.data} -->",end="")
        current_node=current_node.next

  #Insert an element at the beginnning of the Linked List
  def insertAtBegin(self,data):
    new_node=Node(data)
    if self.head==None:
      self.head=new_node
      return
    else:
      new_node.next=self.head
      self.head=new_node

  #Insert an element at a given position in the Linked List
  def insertAtIndex(self,data,index):
    new_node=Node(data)
    current_node=self.head
    position=0
    if position==index:
      self.insertAtBegin(data)
    else:
      while(current_node!=None and position+1!=index):
        position+=1
        current_node=current_node.next
      if(current_node!=None):
        new_node.next=current_node.next
        current_node.next=new_node
      else:
        print("The index is not valid!")

  #Insert an element at the end of the Linked List
  def insertAtEnd(self,data):
    new_node=Node(data)
    if self.head==None:
      self.head=new_node
      return
    current_node=self.head
    while(current_node.next):
      current_node=current_node.next
    current_node.next=new_node

  #Update the data in the node at a particular position
  def updateNode(self,val,index):
    current_node=self.head
    position=0
    if position==index:
      current_node.data=val
    else:
      while(current_node!=None and position!=index):
        current_node=current_node.next
        position+=1
      if current_node!=None:
        current_node.data=val
      else:
        print("The index is not valid!")

  #Remove the first node in the Linked List
  def removeFirstNode(self):
      if self.head==None:
        return
      self.head=self.head.next

  #Remove the last node in the Linked List
  def removeLastNode(self):
      if self.head==None:
        return
      current_node=self.head
      while(current_node.next.next):
        current_node=current_node.next
      current_node.next=None

  #Remove a node from the given position in the Linked List
  def removeAtIndex(self,index):
    if self.head==None:
      return
    current_node=self.head
    position=0
    if position==index:
      self.removeFirstNode()
    else:
      while(current_node!=None and position+1!=index):
        position=position+1
        current_node=current_node.next
      if current_node!=None:
        current_node.next=current_node.next.next
      else:
        print("The index is not valid")

  #Remove the node given the data from the Linked List
  def removeNode(self,val):
    current_node=self.head
    if current_node.data==val:
      self.removeFirstNode()
      return
    while(current_node!=None and current_node.next.data!=val):
      current_node=current_node.next
    if current_node!=None:
        current_node.next=current_node.next.next
    else:
        print("The node does not exist!")

  #Find the size of the linked list
  def sizeOfLinkedList(self):
    size=0
    if self.head!=None:
      current_node=self.head
      while(current_node!=None):
        size=size+1
        current_node=current_node.next
      return size
    else:
      return 0




llist=LinkedList()
llist.insertAtBegin(10)
llist.insertAtIndex(20,1)
llist.insertAtEnd(30)
llist.updateNode(5,0)
llist.updateNode(10,1)
llist.updateNode(15,2)
llist.removeFirstNode()
llist.insertAtBegin("apple")
llist.removeAtIndex(1)
llist.insertAtIndex("grapes",1)
llist.removeLastNode()
llist.insertAtEnd(30)
llist.insertAtEnd(50)
llist.removeNode(30)
llist.printLinkedList()
print(f"\nThe size of the Linked List is {llist.sizeOfLinkedList()}.")
```
